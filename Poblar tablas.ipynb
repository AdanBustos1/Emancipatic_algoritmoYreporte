## Funciones

```python
#Funcion para conectar base de datos

def CargarDataframe(nombreColeccion):
    URI_CONNECTION="mongodb+srv://grupo1tripulaciones:grupo1tripulaciones@cluster0.cm9sh0x.mongodb.net/emancipatic?retryWrites=true&w=majority"
    NOMBRE_BBDD = 'emancipatic'
    cliente = pymongo.MongoClient(URI_CONNECTION)
    baseDatos = cliente[NOMBRE_BBDD]
    coleccion = baseDatos[nombreColeccion]
    dataframe = pd.DataFrame(list(coleccion.find()))
    cliente.close()
    return dataframe

```

```python
#Funcion para obtener los id

def obtener_id(data,nombre_columna,valor):
    """ Esta función sirve para obtener el _id de la fila del dataframe de acuerdo a un valor especifico de una columna dada"""

    data= data.loc[data.loc[:,nombre_columna]== valor]  
    data= data.reset_index()
    return data.at[0,'_id']
```

```python
#funcion fecha aleatoria
def fecha_aleatoria (inicio, fin):
    """ inicio= sera la fecha más antigua
        fin= sera la fecha actual"""
    random_date = inicio + (fin - inicio) * random.random()
    return random_date.isoformat()
```

```python
#funcion fecha hace n años
def fecha_hace_n_anios(anios):
    fecha= datetime.now() + dateutil.relativedelta.relativedelta(years=-anios)
    return fecha
```
